```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

team_details = []

for i in range(1,999):
    url = 'https://www.scrapethissite.com/pages/forms/?page_num={}'.format(i)
    response = requests.get(url)
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    # Find all the team elements
    team_elements = soup.find_all('tr', class_='team')
    if len(team_elements) == 0:
        break
    else:
        for element in team_elements:
            team_name = element.find('td', class_='name').text.strip()
            year = element.find('td', class_='year').text.strip()
            wins = element.find('td', class_='wins').text.strip()
            losses = element.find('td', class_='losses').text.strip()
            ot_losses = element.find('td', class_='ot-losses').text.strip()
    
            # Check if the element exists before accessing its text attribute
            win_perc_element = element.find('td', class_='pct text-success')
            win_perc = win_perc_element.text.strip() if win_perc_element else ''
    
            goal_for = element.find('td', class_='gf').text.strip()
            goal_against = element.find('td', class_='ga').text.strip()
    
            # Check if the element exists before accessing its text attribute
            plus_minus_element = element.find('td', class_='diff text-success')
            plus_minus = plus_minus_element.text.strip() if plus_minus_element else ''
    
            team_details.append({
                'Team Name': team_name,
                'Year': year,
                'Wins': wins,
                'Losses': losses,
                'OT_Losses': ot_losses,
                'Win_perc': win_perc,
                'Goal_for': goal_for,
                'Goal_against': goal_against,
                'plus_minus': plus_minus
            })
    print('Web Scrapping of Page {} done.'.format(i))
        
# Create a DataFrame from the team details
df = pd.DataFrame(team_details)

# Print the DataFrame
df

# Print the scraped team data
# for team in team_details:
#     print('Team Name:', team['team_name'])
#     print('Year:', team['Year'])
#     print('Wins:', team['Wins'])
#     print('Losses:', team['Losses'])
#     print('OT Losses:', team['OT_Losses'])
#     print('Win Percentage:', team['Win_perc'])
#     print('Goals For:', team['Goal_for'])
#     print('Goals Against:', team['Goal_against'])
#     print('Plus/Minus:', team['plus_minus'])
#     print()
```
